```python
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
```

```python
data = pd.read_csv('German_Credit.csv', index_col='ID')
```

```python
data.head()
```

```python
X = data
Y = data['Credit_amount']
```

```python
dummied_col = pd.get_dummies(X['Credit_status'],prefix='Credit_status')
```

```python
X = X.join(dummied_col)
```

```python
X.head()
```

```python
import numpy as np
```

```python
y = np.array(Y)
```

```python
del(X['Credit_amount'])
```

```python
del(X['Credit_status'])
```

```python
lasso = Lasso(alpha=2)
```

```python
lasso.fit(X,Y)
```

```python
lasso.alpha
```

```python
lasso.coef_
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25)
```

```python
lasso.fit(X_train,Y_train)
```

```python
lasso.predict(X_test)
```

```python
lassocv = LassoCV(eps = 0.0001, n_alphas = 5000,
                  cv = 5, n_jobs= -1, normalize = True,
                  max_iter=1000000, verbose=True)
lassocv.fit(X_train,Y_train)
```

```python
lassocv.coef_
```

```python
y_pred = lassocv.predict(X_test)
```

```python
lassocv.alpha_
```

```python
y_pred
```

```python
import math
ans = np.array(y_pred - Y_test)
```

```python
lassocv.score(X_train, Y_train)
```

```python
math.sqrt(sum(ans**2))
```
